```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
consulados = pd.read_csv("consulados.csv")
consulados["CONSULADO"] = consulados["CONSULADO"].astype(str)
participacao = pd.read_csv("participacao_consulados.csv")
participacao["Codigo"] = participacao["Codigo"].astype(str)
residentes_idades = pd.read_csv("entradas.csv")
residentes_idades["MAI_code"] = residentes_idades["MAI_code"].fillna(0).astype(int).astype(str)
MAI_to_ISO = dict(zip(residentes_idades["MAI_code"], residentes_idades["ISO_code"]))
```

```python
consulados
```

```python
grouped_df = participacao.groupby(['Ano', 'Eleição', participacao['Codigo'].str[:4]])

# Aggregating the data for each group
aggregated_df = grouped_df.agg({
    'Pais':"first",
    'Votantes': 'sum',
    'Nulos': 'sum',
    'Inscritos': 'sum',
    'Codigo': 'count'  # Counting the number of electoral constituencies in each group
    
})

# Renaming the columns
aggregated_df = aggregated_df.rename(columns={'Codigo': 'N_Consulados'})

# Resetting the index to turn the grouped columns back into regular columns
#aggregated_df = aggregated_df.set_index(["Codigo", "Ano", "Eleição"])

# Printing the resulting DataFrame
aggregated_df.head()
```

```python
aggregated_df = aggregated_df.reset_index()

aggregated_df["ISO_code"] = aggregated_df["Codigo"].map(MAI_to_ISO)
aggregated_df = aggregated_df.rename(columns={"Codigo": "MAI_code"})

aggregated_df = aggregated_df.set_index(["Ano", "Eleição", "MAI_code", "ISO_code"])

```

```python
grouped_df = consulados.groupby([consulados['CONSULADO'].str[:4]])

# Aggregating the data for each group
aggregated_df = grouped_df.agg({
    'Continente' : "first",
    'Pais':"first",
    'PT_MAI_12' : "sum",
    'PT_MAI_13' : "sum",
    'PT_MAI_14' : "sum",
    'PT_MAI_15' : "sum",
    'PT_MAI_16' : "sum",
    'PT_MAI_17' : "sum",
    'PT_MAI_18' : "sum",
    'PT_MAI_19' : "sum",
    'PT_MAI_20' : "sum",
    'PT_MAI_21' : "sum",
    'CONSULADO': 'count'  # Counting the number of electoral constituencies in each group
    
})

# Renaming the columns
aggregated_df = aggregated_df.rename(columns={'CONSULADO': 'N_Consulados'})

# Resetting the index to turn the grouped columns back into regular columns
#aggregated_df = aggregated_df.set_index(["Codigo", "Ano", "Eleição"])

# Printing the resulting DataFrame
aggregated_df.head()
```

```python
aggregated_df = aggregated_df.reset_index()

aggregated_df["ISO_code"] = aggregated_df["CONSULADO"].map(MAI_to_ISO)
aggregated_df = aggregated_df.rename(columns={"CONSULADO": "MAI_code"})

aggregated_df = aggregated_df.set_index(["MAI_code", "ISO_code"])
```

```python
aggregated_df = aggregated_df.drop(columns=["index"])
```
